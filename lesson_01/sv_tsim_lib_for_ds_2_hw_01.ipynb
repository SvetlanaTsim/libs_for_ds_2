Попробуем определить эффективность маркетинговых каналов по привлечению платящих игроков

Будем работать с полями:

marketing_channel: рекламный канал

converted: конверсия в покупку, 1 - пользователь совершил покупку, 0 - нет

```python
full_df['converted'].value_counts()
```

```python
full_df[full_df['converted'] == True]['marketing_channel']
```

```python
full_df[full_df['converted'] == True].groupby('marketing_channel').count()['user_id']
```

```python
full_df.groupby(['marketing_channel', 'converted']).count()['user_id']
```

```python
full_df.groupby(['marketing_channel', 'converted']).count()['user_id'].unstack()
```

```python
full_df.pivot_table(
    index='marketing_channel',
    columns='converted',
    values='user_id',
    aggfunc='count'
)
```

```python
#теперь попробуем найти процентное соотношение (группировка с нормализацией)
pd.crosstab(index=full_df['marketing_channel'],
            columns=full_df['converted'],
            normalize='index')
```

**Вывод**:

Наиболее эффективный канал в абсолютном выражении: 

House Ads	- 298 покупателей

В процентном соотношении конверсии:

Email	-	0.341593 конверсия

Попробуем определить самую раннюю дату подписки на сервис (поле "date_subscribed")

```python
#преобразуем в datetime

from datetime import datetime, timedelta

full_df['date_subscribed'] = pd.to_datetime(full_df['date_subscribed'])

full_df.head(20)
```

```python
min_date = full_df['date_subscribed'].min()

print(f'Cамая ранняя дата подписки на сервис {min_date}')
```
