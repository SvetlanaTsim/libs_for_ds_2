**Выберите дом клиенту. Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?**

```python
#очень красивый - это будем считать показатели 3 и 4
house_df['view'].value_counts()
```

```python
house_df['waterfront'].value_counts()
```

```python
condition1 = (house_df['waterfront'] == 1) | (house_df['view'] >= 3)
```

```python
house_df[condition1].head()
```

```python
#очень хорошее состояние будем считать 4 и 5
house_df['condition'].value_counts()
```

```python
condition2 = (house_df['condition'] >= 4) & (house_df['yr_built'] >= 1980)
```

```python
house_df[condition2]
```

```python
#варианты подходящих домов
house_df[condition1 & condition2]
```

```python
print(f"У клиента вариантов: {house_df[condition1 & condition2]['id'].count()}")
```

```python
#максимальная цена домов
max_price = house_df[condition1 & condition2]['price'].max()
max_price
```

```python
#минимальная цена домов
min_price = house_df[condition1 & condition2]['price'].min()
min_price 
```

```python
print(f"Ценовой диапазон домов: {min_price} - {max_price}")
```

**Выберите дом клиенту. Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?**

```python
conditon3 = (house_df['floors'] == 2) & (house_df['sqft_basement'] == 0) & (house_df['price'] <= 150000)
```

```python
print(f"Оценка у подходящих домов в среднем: {house_df[conditon3]['grade'].mean()}")
```
